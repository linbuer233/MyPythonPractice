```python
import matplotlib.pyplot as plt 
import numpy as np
import pandas as pd
plt.rcParams['font.sans-serif'] = ['SimHei']  ###防止无法显示中文并设置黑体
plt.rcParams['axes.unicode_minus'] = False  ###用来正常显示负号
```

```python
import os
xmlfile=[]
palfile=[]
for filepath,dirnames,filenames in os.walk(r'path'):
    for filename in filenames:
        a=os.path.join(filepath,filename)
        if '.xml' in a:
            xmlfile.append(a)
        if '.pal' in a:
            palfile.append(a)
```

```python
def createxmlCmap(file):
    import numpy as np
    from matplotlib.colors import ListedColormap
    with open(file) as f:
        # print(f.readline())
        rgbalist=[]
        entry_valuelist=[]
        alllines=f.readlines()
        for i in alllines:
            # print(i)
            if 'palette name' in i:
                palette_name=i.split('"')[1]
                # print(palette_name)
            else:
                palette_name=file.split('\\')[-1].replace('.xml','')
            if 'rgba' in i:
                # print(i.split('"'))
                rgbaindex=i.split('"').index(' rgba=')
                rgba=i.split('"')[rgbaindex+1].split(',')
                rgba=tuple(np.array(rgba[:4],dtype=np.int64)/256)
                rgbalist.append(rgba)
                entry_value=i.split('"')[1]
                entry_valuelist.append(float(entry_value))
    return ListedColormap(rgbalist,name =palette_name,N=len(rgbalist)),entry_valuelist,palette_name
```

```python
def plotcolorbar(file):
    import matplotlib as mpl
    cmap,value,labelname =createxmlCmap(file)
    print(labelname)
    fig, ax = plt.subplots(figsize=(6, 1))
    fig.subplots_adjust(bottom=0.5)
    bounds =sorted(value)
    norm = mpl.colors.BoundaryNorm(bounds, cmap.N+1, extend='both')
    fig.colorbar(mpl.cm.ScalarMappable(norm=norm, cmap=cmap),
                cax=ax, orientation='horizontal',
                label=labelname)
    plt.show()
```

```python
plotcolorbar(xmlfile[1])
plotcolorbar(xmlfile[1])
```

```python
def createpalCmap(file):
    import numpy as np
    from matplotlib.colors import ListedColormap
    try:
        with open(file,encoding='GBK') as f:
            name=f.readline().split(' ')[-1]
        colors=np.loadtxt(file,skiprows=1,encoding='GBK')
    except:
        with open(file) as f:
            name=f.readline().split(' ')[-1]
        colors=np.loadtxt(file,skiprows=1)
    colorlist=[]
    for i in range(len(colors)):
        colorlist.append(tuple(colors[i,:]/256))
    cm=ListedColormap(colorlist,name=name,N=len(colors))
    return cm,name
```

```python
for i in palfile:
    print(i.split('\\')[-1].replace('.pal',''))
    cm,name=createpalCmap(i)
    import matplotlib.pyplot as plt
    import matplotlib as mpl

    fig, ax = plt.subplots(figsize=(6, 1))
    fig.subplots_adjust(bottom=0.5)

    cmap = cm
    bounds = range(0,1000,cmap.N)
    norm = mpl.colors.BoundaryNorm(bounds, cmap.N, extend='both')

    fig.colorbar(mpl.cm.ScalarMappable(norm=norm, cmap=cmap),
                cax=ax, orientation='horizontal', label=name)
    plt.show()
```
