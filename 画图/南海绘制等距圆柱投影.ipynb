```python
import numpy as np
import matplotlib.pyplot as plt
from copy import copy
import cartopy.crs as ccrs
import cartopy.feature as cfeature
from cartopy.io.shapereader import Reader
from cartopy.mpl.ticker import LongitudeFormatter, LatitudeFormatter
```

```python
fig=plt.figure(figsize=(16,9))
ax1=fig.subplots(1,1,subplot_kw={'projection':ccrs.PlateCarree(central_longitude=105)})
# 
box1 = [70, 135, 17.5, 55]
box2 = [105, 125, 0, 25]
xstep=5
ystep=5
ax1.coastlines('50m')
ax1.add_geometries(Reader('D:\\maplist\\China_province\\bou2_4l.shp').geometries(), ccrs.PlateCarree(),
                   facecolor='none', edgecolor='gray', linewidth=0.8)
ax1.set_extent(box1, crs=ccrs.PlateCarree())
# 
ax1.set_xticks(np.arange(box1[0], box1[1] + xstep, xstep), crs=ccrs.PlateCarree())
ax1.set_yticks(np.arange(box1[2], box1[3] + ystep, ystep), crs=ccrs.PlateCarree())
# 0EW
lon_formatter = LongitudeFormatter(zero_direction_label=False)
lat_formatter = LatitudeFormatter()
ax1.xaxis.set_major_formatter(lon_formatter)
ax1.yaxis.set_major_formatter(lat_formatter)

# 
ax2 = fig.add_axes([0.76, 0.09, 0.1, 0.3], projection=ccrs.PlateCarree())
ax2.coastlines('50m')
ax2.set_extent(box2, crs=ccrs.PlateCarree())
ax2.add_geometries(Reader('D:\\maplist\\China_province\\bou2_4l.shp').geometries(), ccrs.PlateCarree(),
                   facecolor='none', edgecolor='gray', linewidth=0.8)
```
