```python
T = ds1["t"][0, :, :, :]
T=T+273.16
T
```

```python
def TDcompute(RH, T):
    import numpy as np
    es = np.zeros((RH.shape[0], RH.shape[1], RH.shape[2]))
    Td = np.zeros((RH.shape[0], RH.shape[1], RH.shape[2]))
    Ttemp = T
    a = 17.2693882
    b = 35.86
    es = 6.1078 * np.exp(a * (T - 273.16) / (T - b))
    e = es * RH / 100
    for t_i in range(RH.shape[0]):
        for i in range(RH.shape[1]):
            for j in range(RH.shape[2]):
                while e[t_i, i, j] < es[t_i, i, j]:
                    Ttemp[t_i, i, j] = Ttemp[t_i, i, j] - 0.05
                    es[t_i, i, j] = 6.1078 * \
                        np.exp(a * (Ttemp[t_i, i, j] -
                               273.16) / (Ttemp[t_i, i, j] - b))
                Td[t_i, i, j] = Ttemp[t_i, i, j]
    return Td
```

```python
%time TDcompute(RH.data,T.data)
```

```python
from dask import delayed
z=delayed(TDcompute)(RH.data,T.data)
%time z.compute()
```
