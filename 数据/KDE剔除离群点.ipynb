```python
import numpy as np
from scipy.stats import gaussian_kde

# 准备数据
data = np.random.multivariate_normal(mean=[0, 0], cov=[[1, 0.5], [0.5, 1]], size=1000)  # 示例数据，可以替换为你的实际数据

# 初始化KDE模型
kde = gaussian_kde(data.T)  # 注意要将数据转置

# 计算概率密度估计值
density = kde.evaluate(data.T)

# 指定前百分数
percentile = 0.99  # 替换为你要计算的百分数值

# 计算概率密度的阈值
threshold = np.quantile(density, 1 - percentile)

# 筛选出不是离群点的数据
filtered_data = data[density >= threshold]

# 打印筛选后的数据
print(f"Filtered data points:\n{filtered_data}")
```
