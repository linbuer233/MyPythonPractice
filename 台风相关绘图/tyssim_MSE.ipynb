```python
import numpy as np
import xarray as xr
from sklearn.metrics import mean_squared_error
```

```python
cma=xr.open_dataset('track0xr_cma.nc')
cma
```

```python
det=xr.open_dataset('track0xr_det.nc')
det
```

```python
# 年份相加
sumcma=cma.sum(dim='year')
sumdet=det.sum(dim='year')
```

```python
# MSE
MSEcma=((sumcma-sumdet)**2).mean()['__xarray_dataarray_variable__'].data
MSEcma
```

```python
# R quared
R=1-mean_squared_error(sumcma['__xarray_dataarray_variable__'].data,sumdet['__xarray_dataarray_variable__'].data)/np.var(sumcma['__xarray_dataarray_variable__'].data)
R
```
