```python
def RGB_to_Hex(rgb):
    RGB = rgb          # 将RGB格式划分开来
    color = '#'
    for i in RGB:
        num = int(i)
        # 将R、G、B分别转化为16进制拼接转换并大写  hex() 函数用于将10进制整数转换成16进制，以字符串形式表示
        color += str(hex(num))[-2:].replace('x', '0').upper()
    print(color)
    return color
```

```python
RGB_to_Hex(img[a,20,:][0][0])
```
