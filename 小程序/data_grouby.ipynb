
# 利用groupby对数据进行合并操作，常用于对时间序列合并，例如对时间间隔是日的求月平均或者对几年求年平均

```python
import numpy as np
import xarray as xr
import pandas as pd
```

```python
# 读取数据
# dataframe形式的数据
df =pd.read_csv('../pythonhomework6csv.csv')
# dataset/dataarray形式的数据
ds =xr.open_dataset("D:/WORKcode/DuanQiQiHou/data/hgt.mon.mean.nc")
```

```python
#例如对2015-2020年的每年求月平均，可以这样
dfg=df.loc[df['year'].isin(np.arange(2015,2021,1))].groupby(['year','month']).mean()
dfg=dfg.reset_index()
dfg
```

```python
# 对于dataset来说（xarray）1970-2010

hgt=ds['hgt'].groupby(ds.time.dt.year).mean(dim='time').loc['19700101':'20100101',:,:,:]
hgt
```

```python
# 当然你也可以通过这样去实现上述的操作，不过就比较麻烦，平时求平均值用下面这个比较多
hgt1=ds['hgt'].loc['2010',:,:,:]
hgt1.mean(dim='time')
```
